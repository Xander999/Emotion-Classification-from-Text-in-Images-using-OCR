#  OCR Preprocessing For Better Resuts

```python
from PIL import Image
import pytesseract as pt
import argparse
import cv2

pt.pytesseract.tesseract_cmd = r"/usr/bin/tesseract"
```

```python
import os
arr = os.listdir('Data Files/Dataset/')
print(arr[:10])
```

```python
image=cv2.imread('Data Files/Dataset/Test125.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = cv2.threshold(gray, 0, 255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
gray = cv2.medianBlur(gray, 3)
filename = "{}.png".format(os.getpid())
cv2.imwrite(filename, gray)
print(filename)
text = pt.image_to_string(image)
print(text)
```

# Test-1Extracting the test from image using thresholding and blurring   

```python
nostr=[]
yesstr=[]
preprocessnostr=[]
preprocessyesstr=[]
thryes=[]
blryes=[]

for x in arr:
    img=cv2.imread('Data Files/Dataset/'+x)
    txt=pt.image_to_string(img)
    print('File Name :',x)
    if(txt==''):
        nostr.append(x)
        gry=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        thr=cv2.threshold(gry, 0, 255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
        blr=cv2.medianBlur(gry, 3)
        
        txt1=pt.image_to_string(thr)
        txt2=pt.image_to_string(blr)
        
        if(txt1!=''):
            print('Recovered using Thresholding :',txt1)
            preprocessyesstr.append(x)
            thryes.append(x)
        elif(txt2!=''):
            print('Recovered using Blurring :',txt2)
            preprocessyesstr.append(x)
            blryes.append(x)
        else:
            preprocessnostr.append(x)
            
        
    else:
        yesstr.append(x)
```

```python
print('No Message in First Order :',len(nostr))
print('Message in First Order :',len(yesstr))
print('---------------------------------------------------------')
print('No Message in Second Order :',len(preprocessnostr))
print('Message in Second Order :',len(preprocessyesstr))
print('---------------------------------------------------------')
print('Recovered using Thresolding :',len(thryes))
print('Recovered using Blurring :',len(blryes))
```

# Test 2: Increasing the Contrast by 1.5 times over application of Test 1

```python
nostr=[]
yesstr=[]
preprocessnostr=[]
preprocessyesstr=[]
thryes=[]
blryes=[]

#making some pre-processing in the images such as increasing the contrast

for x in arr:
    oimg=cv2.imread('Data Files/Dataset/'+x)
    img = cv2.convertScaleAbs(oimg, alpha=1.5, beta=0)
    txt=pt.image_to_string(img)
    print('File Name :',x)
    if(txt==''):
        nostr.append(x)
        gry=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        thr=cv2.threshold(gry, 0, 255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
        blr=cv2.medianBlur(gry, 3)
        
        txt1=pt.image_to_string(thr)
        txt2=pt.image_to_string(blr)
        
        if(txt1!=''):
            print('Recovered using Thresholding :',txt1)
            preprocessyesstr.append(x)
            thryes.append(x)
        elif(txt2!=''):
            print('Recovered using Blurring :',txt2)
            preprocessyesstr.append(x)
            blryes.append(x)
        else:
            preprocessnostr.append(x)
            
        
    else:
        yesstr.append(x)
```

```python
print('No Message in First Order :',len(nostr))
print('Message in First Order :',len(yesstr))
print('---------------------------------------------------------')
print('No Message in Second Order :',len(preprocessnostr))
print('Message in Second Order :',len(preprocessyesstr))
print('---------------------------------------------------------')
print('Recovered using Thresolding :',len(thryes))
print('Recovered using Blurring :',len(blryes))
```
