# Experiments from image processing file 1

```python
from PIL import Image
import pytesseract as pt
import argparse
import cv2

pt.pytesseract.tesseract_cmd = r"/usr/bin/tesseract"
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
arr = os.listdir('Data Files/Dataset/')
print(arr[:10])
```

# Extracting text from image

```python
nostr=[]
yesstr=[]
preprocessnostr=[]
preprocessyesstr=[]
thryes=[]
blryes=[]

text=[]

#making some pre-processing in the images such as increasing the contrast

for x in arr:
    img1=cv2.imread('Data Files/Dataset/'+x)
    gry=cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
    im = cv2.filter2D(gry, -1, kernel)
    txt=''
    txt=pt.image_to_string(im)
    print('File Name :',x)
    if(txt==''):
        nostr.append(x)
        thr=cv2.threshold(gry, 0, 255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
        blr=cv2.medianBlur(gry, 3)
        
        txt1=pt.image_to_string(thr)
        txt2=pt.image_to_string(blr)
        
        if(txt1!=''):
            print('Recovered using Thresholding :',txt1)
            preprocessyesstr.append(x)
            txt=txt.strip()
            text.append(txt)
            thryes.append(x)
        elif(txt2!=''):
            print('Recovered using Blurring :',txt2)
            preprocessyesstr.append(x)
            txt=txt.strip()
            text.append(txt)
            blryes.append(x)
        else:
            preprocessnostr.append(x)
            text.append(' ')
            
        
    else:
        yesstr.append(x)
        txt=txt.strip()
        text.append(txt)
```

```python
data=pd.DataFrame({'fileName':arr, 'text':text})
data.to_csv('MainFile.csv')
```
